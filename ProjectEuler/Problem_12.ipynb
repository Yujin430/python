```python
def min_tri(n):
    L= [((k*(k+1))/2) for k in range(n*n)]
    for i in L:
        count = 0
        for j in range(i,0,-1):
            if(i%j==0):
                if(j>(i/j)):
                    count+=2
                elif(j==(i/j)):
                    count+=1
                else:
                    break
            if(count>=n):
                return i
print min_tri(4)
#되긴 되는데 너무 느리고 150부터도 겔겔거림.
```
