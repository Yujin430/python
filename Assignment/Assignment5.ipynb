> False  
> \>\>\> print d < 10  
> True  
> \>\>\> print c == 17  
> True  
> \>\>\> print d != 9  
> False 

```python
class Counter():
    #요구사항1
    def __init__(self,count,step=1):
        self.count = count
        self.step = step
    #요구사항2
    def __str__(self):
        return str('[Count (Step: %d)] %d' % (self.step, self.count))
    #요구사항3
    def incr(self):
        self.count += self.step
    #요구사항4
    def __call__(self, *args):
        self.incr()
    #요구사항5
    def __add__(self, num):
        self.count+=num
        return self
    def __sub__(self, num):
        self.count-=num
        return self
    #요구사항6
    def __gt__(self, num):
        return self.count> num
    def __lt__(self,num):
        return self.count< num
    def __eq__(self,num):
        return self.count==num
    def __ne__(self,num):
        return self.count!=num
#요구사항1    
c = Counter(10)
d = Counter(10,2)
#요구사항2
print c
print d
print
#요구사항3
c.incr()
d.incr()
print c
print d
print
#요구사항4
c()
d()
print c
print d
print
#요구사항5
c= c + 5
d= d -5
print c
print d
print
#요구사항6
print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

- 생성자를 만들 때는 \_\_init\_\_을 재정의 하면 됩니다. count와 step을 클래스가 받을 수 있도록 인자로 만들어 주는데 그 때 step은 디폴트 값으로 1을 주는 것을 잊지 않으면 됩니다.
- 출력해주는 메소드를 구현할때는 문자열 포매팅을 이용해서 count 값과 step값을 받아올 수 있도록 해주었습니다.
- incr메소드에서는 카운트 값을 스텝만큼 증가시켜 주었습니다.
- 인스턴스 객체를 직접 호출할 수 있게 하기 위해 \_\_call\_\_을 재정의 해주었습니다. 그 메소드 내부에 self.incr()을 써 주어 call이 불릴 때 incr도 불리도록 만들어주었습니다.
- +나 - 연산을 위해서 \_\_add\_\_와 \_\_sub\_\_을 재정의 해줍니다. 카운트에서 인자로 들어오는 수를 빼고 자신을 리턴하게 하면 print문에서도 제대로 출력됩니다.

#### 04) 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

- \_\_init\_\_() 메소드는 생성자 메소드입니다. 생성할 때 인자로 받은 리스트의 원소들을 원소로 가지는 리스트를 만들고, eliminate_duplicate 함수를 클래스를 이용하여 호출해줍니다.
- \_\_str\_\_() 메소드는 출력을 할 때 이용되는 함수입니다. 마지막에 반환해줄 문자열을 일단 앞부분 Myset: { 을 적어서 만들고, 원소들을 문자열로 바꾸어 반환해줄 문자열에 추가시켜주고, 마지막에는 ,가 필요 없으므로 슬라이싱을 이용하여 문자열의 마지막원소까지의 문자열에 뒷부분 } 를 추가시켜 준 후, 그 문자열을 반환해줍니다.
- eliminate_duplicate()함수는 중복되는 원소들을 제거해주는 함수입니다. 먼저 빈 리스트(s)를 하나 만들어주고, 인자로 들어온 리스트(l)의 원소가 만들어준 리스트(s)에 없으면 집어넣어줍니다. 그리고 인자로 들어온 리스트(l)를 비워주고, 다시 인자로 들어왔던 리스트(l)에 만들어 두었던 리스트(s)의 원소를 차례로 집어넣어줍니다. 그러면 중복된 원소가 제거된 리스트를 가질 수 있게 됩니다.

#### 05) 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

> \>\>\> u = s | t   
> \>\>\> print u  
> MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}  

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

> \>\>\> u = s & t   
> \>\>\> print u  
> MySet: {2, 3}  

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

> \>\>\> s = MySet([1, 2, 3])  
> \>\>\> t = MySet([3, 4, 5])  
> \>\>\> u = s - t    
> \>\>\> print u  
> MySet: {1}  


```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    #요구사항 1
    def __or__(self,other):
        s=[]
        for e in self:
            if e not in s:
                s.append(e)
        for e in other:
            if e not in s:
                s.append(e)
        s.sort()
        return MySet(s)
    #요구사항 2
    def __and__(self,other):
        s=[]
        for e in self:
            if e in other:
                if e not in s:
                    s.append(e)
        s.sort()
        return MySet(s)
    #요구사항 3
    def __sub__(self,other):
        s=[]
        for e in self:
            if e not in other:
                if e not in s:
                    s.append(e)
        s.sort()
        return MySet(s)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
print
#요구사항1
u = s | t
print u
#요구사항2
u = s & t
print u
#요구사항3
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s - t
print u
```

- 요구사항 1, 2, 3을 해결하기 위해 각각 \_\_or\_\_, \_\_and\_\_, \_\_sub\_\_메소드를 재정의해주었습니다. or의 경우 self와 other에 존재하는 원소 모두를 추가시키는 중복되지 않게 추가하는 식으로 진행했고 교집합의 경우 self와 other 둘 다에 존재하는 원소만 중복되지 않게 추가하였고, 차집합의 경우 self에는 있지만 other에는 없는 원소만 추가시키게 하였습니다.

  - 예를 들어 342를 영어로 쓰면 three hundred and forty-two 가 되어서 23 글자,
  - 115 = one hundred and fifteen 의 경우에는 20 글자가 됩니다.

```python
#일의자리 수를 처리하는 함수, 백의자리수도 처리 가능
def units_digit(n):
    if(n==0):
        return 0
    if(n==1):
        return len('one')
    elif(n==2):
        return len('two')
    elif(n==3):
        return len('three')
    elif(n==4):
        return len('four')
    elif(n==5):
        return len('five')
    elif(n==6):
        return len('six')
    elif(n==7):
        return len('seven')
    elif(n==8):
        return len('eight')
    else:
        return len('nine')
#십의자리 수를 처리하는 함수
def tens_digit(n):
    if(n==0):
        return 0
    if(n==2):
        return len('twenty')
    elif(n==3):
        return len('thirty')
    elif(n==4):
        return len('forty')
    elif(n==5):
        return len('fifty')
    elif(n==6):
        return len('sixty')
    elif(n==7):
        return len('seventy')
    elif(n==8):
        return len('eighty')
    else:
        return len('ninety')
#10~19를 처리하는 함수
def ten_to_nineteen(n):
    if(n==0):
        return len('ten')
    elif(n==1):
        return len('eleven')
    elif(n==2):
        return len('twelve')
    elif(n==3):
        return len('thirteen')
    elif(n==4):
        return len('fourteen')
    elif(n==5):
        return len('fifteen')
    elif(n==6):
        return len('sixteen')
    elif(n==7):
        return len('seventeen')
    elif(n==8):
        return len('eighteen')
    else:
        return len('nineteen')
    
NUM=1000
length_of_word = 0
for i in range(1,NUM+1):
    if(len(str(i))==1):#한자리 수
        length_of_word+=units_digit(i)#일의 자리 수 처리
    elif(len(str(i))==2):#두자리 수
        if(int(str(i)[0])==1):#십의자리 수가 1이면,
            length_of_word+=ten_to_nineteen(int(str(i)[1]))#10~19까지의 단어 처리
        else:#십의자리 수가 1이 아니면,
            length_of_word+=tens_digit(int(str(i)[0]))#십의자리 수 처리
            length_of_word+=units_digit(int(str(i)[1]))#일의자리 수 처리
    elif(len(str(i))==3):#세자리 수
        if(int(str(i)[1])==1):#십의자리 수가 1이면
            length_of_word+=units_digit(int(str(i)[0]))#백의자리 수 처리
            length_of_word+=len('hundred'+'and')#hundred와 and의 길이를 더해줌
            length_of_word+=ten_to_nineteen(int(str(i)[2]))#10~19까지의 단어 처리
        else:#십의자리 수가 1이 아니면
            if(int(str(i)[1])==0 and int(str(i)[2])==0):#십의자리,일의자리가 모두 0일때 
                length_of_word+=units_digit(int(str(i)[0]))#백의자리 수 처리
                length_of_word+=len('hundred')#hundred의 길이를 더해줌
            else:#십의자리나 일의자리중 하나라도 0이 아닐때,
                length_of_word+=units_digit(int(str(i)[0]))#백의자리 수 처리
                length_of_word+=len('hundred'+'and')#hundred와 and의 길이를 더해줌
                length_of_word+=tens_digit(int(str(i)[1]))#십의자리 수 처리
                length_of_word+=units_digit(int(str(i)[2]))#일의자리 수 처리
    else:#네자리 수(1000)
        length_of_word+=len('one'+'thousand')#one과 thousand의 길이를 더해줌
        
print length_of_word
```

- 이 문제는 자세히 보면 케이스 분류만 제대로 해 주면 되는 문제입니다. 일의 자리 수와 백의 자리수를 처리 가능한 함수, 십의 자리 수를 처리 가능한 함수, 10~19를 처리 가능한 함수를 각각 만들어주고, 주어진 범위(1~1000)의 반복을 돌려 경우에 따라 함수를 호출하여 영어로 된 숫자를 얻어내고 그것의 길이를 마지막에 반환할 length_of_word에 더해주는 식으로 문제를 해결하였습니다.

- a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다.


- 예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.
- 또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.
- 따라서 220과 284는 친화쌍이 됩니다.


- 10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.

```python
import math
def Divisor_SUM(MAX):
    divisor_sum = []
    for i in range(0,MAX+1):
        divisor_sum.append(0)
    for i in range(2,MAX+1):#만 이하의 수들에 대하여
        temp = (int)(math.sqrt(i))
        for j in range(temp,0,-1):
            if(i%j==0):
                if(j<(i/j)):
                    divisor_sum[i]+=j
                    if(j!=1):
                        divisor_sum[i]+=i/j
                elif(j==(i/j)):
                    divisor_sum[i]+=j
                else:
                    break
    return divisor_sum
Max=10000
arr = Divisor_SUM(Max)
FriendNUM=[]
for num in arr:
    if(num==1 or num in FriendNUM):
        continue
    if(num<=10000 and arr[num]<=10000 and arr[arr[num]]==num):
        if(arr[num]!=num):
            FriendNUM.append(num)
            FriendNUM.append(arr[num])
#print FriendNUM
print sum(FriendNUM)
```

```python
import urllib2
import string

def tag_delete(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:]," ")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:]," ")
    
    return string

#자바스크립트 제거
def javascript_delete(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:]," ")
    return string

#css 제거
def stylesheet_delete(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:]," ")
    return string

#구두문자 제거 함수
def punc_delete(source):
    punctuation = string.punctuation
    for i in range (len(punctuation)):
        source= source.replace(punctuation[i]," ")
        
    return source

#불용어 제거 함수
def stopword_delete(words):
    stopword = [ 'A','a', 'and', 'are', 'as', 'at', 'be', 'but','But', 'by', 'for', 'if','If', 'in', 'into', 'is', 'it','It', 'no','No', 'not','Not', 'of', 'on', 'or', 's', 'such', 't', 'that','That', 'the','The', 'their','Their', 'then','Then', 'there','There', 'these','These', 'they','They', 'this','This', 'to', 'was', 'will', 'with']
    for i in range (len(stopword)):
        for j in range(len(words)):
            if words[j] == stopword[i]:
                words[j] = " "


#사전형태로 출현빈도 저장 함수
def word_count(words):
    words_dict = {}
    for i in words:
        if words_dict.has_key(i) is False:
            words_dict[i]=1
        else:
            words_dict[i]+=1
    return words_dict

def MakeDic(source):
    source=stylesheet_delete(source)
    source=javascript_delete(source)
    source=tag_delete(source)
    source=punc_delete(source)
    words=source.split()
    stopword_delete(words)
    source=" ".join(words)
    words=source.split()
    words=word_count(words)
    return words

class WebWordsFrequency():
    def __init__(self, *arg):
        self.URL = []
        for url in arg:
            self.URL.append(url)
    def addUrl(self, url):
        self.URL.append(url)
    def removeUrl(self, url):
        if url in self.URL:
            self.URL.remove(url)
    def listUrls(self):
        for url in self.URL:
            print url
    def getWordsFrequency(self):
        combine={}
        temp = []
        count = 0
        for url in self.URL:
            temp = MakeDic(urllib2.urlopen(url).read())
            count=1
            if(count==0):
                for i in temp:
                    combine[i]=temp[i]
            else:
                for i in temp:
                    if combine.has_key(i):
                        combine[i]+=temp[i]
                    else:
                        combine[i]=temp[i]
        return combine
    def getMaxFrequencyWords(self):
        combine = self.getWordsFrequency()
        g = combine.items()
        g.sort(key=lambda x:x[1],reverse=True)
        wordlist = []
        for i in range(len(g)-1):
            wordlist.append(g[i])
            if(g[i][1]!=g[i+1][1]):
                break
        if(len(wordlist)==0):
            return None
        return wordlist
#2확인
w1 = WebWordsFrequency('http://www.ebay.com/', 'http://www.bestbuy.com/', 'https://www.amazon.com')
w2 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com')
w3 = WebWordsFrequency()
#3확인
w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')
#4확인
w1.removeUrl('http://www.bestbuy.com/')
w2.removeUrl('http://stackoverflow.com')
#5확인
w1.listUrls()
#6확인
print w1.getWordsFrequency() 
#7확인
print w1.getMaxFrequencyWords()
print w2.getMaxFrequencyWords()
```

####14) 13번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
- 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
- getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.

  - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
  - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.

> \>\>\> w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com')

> \>\>\> w4.getWordsFrequency()
[('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

> \>\>\> w4.getWordsFrequency(reverse=true)
[('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self, reverse=False):
        combine=WebWordsFrequency.getWordsFrequency(self)
        g = combine.items()
        if(reverse==False):
            g.sort(key=lambda x:x[1],reverse=True)
        else:
            g.sort(key=lambda x:x[1],reverse=False)
        return g
w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
print w4.getWordsFrequency()
print
print w4.getWordsFrequency(reverse=True)
```

- 상속받은 클래스 내에서 부모 클래스의 getWordFrequency를 사용하기 위하여 WebWordsFrequency.getWordFrequency(self)와 같은 문장을 사용해주었습니다. reverse인자의 값에 따라 정렬하는 과정에서 먼저 reverse인자 값에 따라 조건을 나누어 준뒤에는 부모 클래스의 getMaxFrequenyWords를 만들 때 이용하였던 방법을 사용하였습니다. 정렬할때 reverse값이 True로 들어오면 False를, False로 들어오면 True를 주어야 합니다.

####15) 다음과 같은 코딩이 가능하도록 OrderedWebWordsFrequency 안에 반복자와 관련된 메소드를 추가하시오.

> \>\>\> for i in w4:  
> \>\>\>       print i  
> ('site, 12)  
> ('science', 11)  
> ('hello', 8)  
> ('world', 2)  
> ('program', 1)  
> ('python', 1)  

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self, reverse=False):
        combine=WebWordsFrequency.getWordsFrequency(self)
        g = combine.items()
        if(reverse==False):
            g.sort(key=lambda x:x[1],reverse=True)
        else:
            g.sort(key=lambda x:x[1],reverse=False)
        return g
    def __init__(self, *arg):
        WebWordsFrequency.__init__(self, *arg)
        self.current_idx=0
        self.li=self.getWordsFrequency(False)
    def __iter__(self):
        return self
    def next(self):
        if self.current_idx>=len(self.li):
            self.current_idx=0
            raise StopIteration
        item = self.li[self.current_idx]
        self.current_idx+=1
        return item
w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
for i in w4:
    print i
```

- next함수를 정의할 때, 인덱스 범위를 넘어서면 raise StopIteration 를 써 주어 예외를 발생시켜줍니다. 예외가 발생하는 경우가 아닐 때에는, 새롭게 정의한 생성자에서 만든 멤버 current_idx와 li를 이용하여 반복자의 역할을 할 수 있도록 만들어 주었습니다.
