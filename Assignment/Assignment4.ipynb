```python
# 01)의 (2)
f=open('s.txt','r')
line = f.read()#파일 내용 읽어오기
f.close()

secondword = []#두 번째 단어를 가질 리스트 secondword
word = line.split()
for i in range(len(word)):
    if(i%2!=0):
        secondword.append(word[i])
secondword.sort()#두 번째 단어를 정렬


linelist = []#파일에 쓸 문자열을 가질 리스트 linelist
mergeword= ''#쪼개어진 단어들을 이어붙인 문자열을 담을 문자열

for i in range(len(secondword)):
    for w in range(len(word)):
        if((w%2!=0) and (secondword[i]==word[w])):
            #두번째 단어와 word의 단어중 뒷단어에 해당하는것을 비교
            mergeword=word[w-1]+' '+word[w]+'\n'
            linelist.append(mergeword)
            #합쳐진 문자열을 linelist에 집어넣음

f = open('s2.txt','w')
f.writelines(linelist)
f.close()

#확인을 위한 출력
f=open('s2.txt','r')
print f.read()
f.close()
```

```python
# 01)의 (3)

#불러오는 과정
f=open('s.txt','r')
s = f.read()
f.close()

word = s.split()#단어를 전부 쪼갬
wordline = []#파일에 쓸 문자열을 가지고 있을 리스트 wordline
temp = ''#파일에 쓸 문자열을 만들 때 사용할 문자열 temp

#파일에 쓸 문자열을 가지고 있을 리스트 wordline을 만드는 과정.
for i in range(0,len(word),3):
    temp = word[i]
    for j in range(i+1,i+3):
        if(j>=len(word)):
            break#만약에 j가 인덱스 범위를 벗어나면 멈춤
        temp = temp+' '+word[j]#단어를 이어 문자열을 만드는 과정
    temp+='\n'#문자열의 끝에 줄바꿈문자 삽입
    wordline.append(temp)#완성된 문자열을 wordline에 추가

f = open('s3.txt','w')
f.writelines(wordline)
f.close()

#확인을 위한 출력
f=open('s3.txt','r')
print f.read()
f.close()
```

####03) range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

(1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.

- range(stop)
- range(start, stop[,step])

(2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

(3) frange 사용 예

- 3-1) frange(0.5)
  - [0.0, 0.1, 0.2, 0.3, 0.4]
- 3-2) frange(1.0, 2.0)
  - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
- 3-3) frange(2.2, 4.0, 0.5)
  - [2.2, 2.7, 3.2, 3.7]

```python
def frange(start,stop=None,step=0.1):
    if(stop==None):
        stop=start+0.0
        start=0.0
        
    L = []
    while(True):
        num=start+len(L)*step
        if((step>0)and(num>=stop)):
            break
        elif((step<0)and(num<=stop)):
            break
        num = round(num,1)
        L.append(num)
    return L
print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

- 들어오는 인자의 갯수가 달라지므로 디폴트 매개변수를 주었고, 특히 step의 경우 디폴트 매개변수를 0.1로 주었습니다. 인자가 하나 들어온 경우에는 stop에 None이 할당됩니다. 그러나 원래 인자가 하나 들어온 경우 그 값은 start가 아닌 stop에 할당 되어야 옳습니다. 따라서 그것을 보정해주는 과정이 필요한데, 그것을 위해 함수의 시작 부분에 조건문을 주어 만약 stop이 None일 경우 stop은 start에 할당되어있던 값을 가지게 하고 start는 0을 주었습니다.
- 그 후 num을 계산하여 리스트에 넣어가는 반복을 진행하였는데, len(L)의 경우 원소의 갯수로, 여기서는 일종의 인덱스와 비슷한 기능을 하게 하였습니다(step과 곱해 얼마가 더해져야 하는지를 나타냄). 그 후, num이 범위 내일때만 올바르게 리스트에 추가하기 위하여 step이 양수일경우 num이 stop보다 크면 멈추고 step이 음수일 경우 num이 stop보다 작으면 멈추었습니다. 그 후 소수 출력이 0.30000000004와 같이 되는 것을 방지하기 위하여 num에 round함수를 취해 반올림해주고 그 후 num을 리스트에 추가하는 식으로 진행한 후 마지막에 리스트를 반환해주는 것으로 마무리했습니다.

####05) 여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.
(1) 다음에 제시되는 함수들을 모두 이용해야 한다.

- split
- map
- join

(2) myinitial() 함수 사용 예

- myinitial("as soon as possible")
  - ASAP

```python
def f(x):
    return x.upper() 

def myinitial(s):
    li = []
    li = s.split()#split 사용
    li_first = []
    for i in range(len(li)):
        li_first.append(li[i][0:1])
    li_upper = map(f,li_first)#map 사용
    result = ''.join(li_upper)#join 사용
    return result
    
print myinitial("as soon as possible")
```

- f함수는 map 함수에서 사용하기 위한 것으로, 받아온 문자를 대문자로 바꾸어 반환해줍니다.
- 일단 split을 사용하여 주어진 문자열을 단어단위로 나누고, 슬라이싱을 통해 가장 앞 글자를 추출하여 그것을 새로운 리스트인 li_first에 모았습니다. 그 후 map함수를 통해 li_first의 원소들에 upper가 적용된 원소들을 li_upper에 모아주고, 마지막으로 그들을 모두 join을 이용해 이어붙인 문자열 result를 구하여 그것을 반환해주는 것으로 마무리했습니다.

####06) 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if(n==0):
        return 1
    if(n==1):
        return 1
    else:
        return n*myfact(n-1)
print myfact(976)
```

- 1부터 n까지의 자연수를 차례로 더하여 구해진 값을 삼각수라고 합니다.
- 예를 들어 7번째 삼각수는 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28이 됩니다.
- 이런 식으로 삼각수를 구해 나가면 다음과 같습니다.

  - 1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...
  
- 이 삼각수들의 약수를 구해봅시다.

  -  1: 1
  -  3: 1, 3
  -  6: 1, 2, 3, 6
  - 10: 1, 2, 5, 10
  - 15: 1, 3, 5, 15
  - 21: 1, 3, 7, 21
  - 28: 1, 2, 4, 7, 14, 28

- 위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.

- 그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?

```python
import math
def min_tri(n):
    L= [((k*(k+1))/2) for k in range(n*n)]
    for i in L:
        count = 0
        temp = (int)(math.sqrt(i))
        for j in range(temp,0,-1):
            if(i%j==0):
                if(j<(i/j)):
                    count+=2
                elif(j==(i/j)):
                    count+=1
                else:
                    break
            if(count>=n):
                return i
print min_tri(500)
```

- 삼각수는 1부터 n까지의 수를 더한 합이므로 n*(n+1)/2 의 공식으로 삼각수의 리스트를 만들 수 있었습니다. 그 후 그렇게 만들어진 리스트를 이용해서 약수의 개수를 세어 가다가 약수의 개수가 500 이상일 때 값을 반환해주는 함수 min_tri를 만들었습니다. 약수의 개수를 구할 때에는 범위를 루트 n 까지로 주어 약수가 하나 나올 때 마다 2씩 더해주고 만약 약수가 루트 n이면 1만 더해주는 방식으로 약수를 구하게 했습니다. 만약 약수를 구하는 도중에 j가 i/j보다 커지게 되면 약수를 다 구한 상태가 되므로 break해 주었습니다. 이렇게 하면 min_tri는 500개 이상의 약수를 갖는 가장 작은 삼각수가 됩니다.

- 양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

  - n → n / 2 (n이 짝수일 때)
  - n → 3 n + 1 (n이 홀수일 때)

- 아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 
- (역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

- 참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.

```python
max_count=0#가장 긴 과정이 얼마나 걸리는지를 저장
max_num=0#구하는 숫자
for i in range(1,1000001):
    n=i
    count=0
    while(n!=1):
        if(n%2==0):
            n=n/2
        else:
            n=3*n+1
        count+=1
    if(count>max_count):
        max_count=count
        max_num=i
print max_num
```

- 영어 불용어 제거 함수인 eng_delete는 이번에 새롭게 만들게 된 함수입니다. 사전객체를 불러와 만약 그 안에 불용어가 있다면 그 항목을 지우는 방식으로 함수를 만들어 주었고, 그 결과 위 블럭의 결과와 현재 블럭의 결과를 비교해보면 위 블록의 결과에는 불용어가 담겨있지만 현재 블럭에는 담겨있지 않다는 것을 알 수 있습니다. 따라서 14-(1)이 해결되었습니다.

- 이후에 사전이 잘 들어갔는지 확인해보기 위해서 한 파일만 직접 로드하여 체크해본 출력결과가 위와같이 아주 잘 되어있습니다. 이렇게 14-(3)도 해결을 하게 되었습니다.

```python
#위 블럭까지의 과정을 토대로 (4)는 저절로 해결이 됨.
#(5) 해결
DictList =[]
for i in range(len(URL_DICT)):
    f = open(URL_DICT[i],'r')
    DictList.append(pickle.load(f))
    #pickle 모듈로 저장한 5개 이상의 words.frequency를 
    #다시 5개 이상의 사전 객체로 로드 하는 코드(로드 후 DictList에 append)
    f.close()
    
MAXWORD=[(None,0),(None,0),(None,0)]

for i in range(len(DictList)):
    for j in DictList[i].items():
        if(j[1]>MAXWORD[0][1]):
            MAXWORD[2]=MAXWORD[1]
            MAXWORD[1]=MAXWORD[0]
            MAXWORD[0]=j
        elif(j[1]>MAXWORD[1][1]):
            MAXWORD[2]=MAXWORD[1]
            MAXWORD[1]=j
        elif(j[1]>MAXWORD[2][1]):
            MAXWORD[2]=j
        else:
            pass
        
print MAXWORD

#이게 잘못나온 것 처럼 보이겠지만 놀랍게도 세 단어가 모두 각각 99번씩 나왔습니다......
```
