```python
import urllib2
import string

def tag_delete(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:]," ")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:]," ")
    
    return string

#자바스크립트 제거
def javascript_delete(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:]," ")
    return string

#css 제거
def stylesheet_delete(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:]," ")
    return string

#구두문자 제거 함수
def punc_delete(source):
    punctuation = string.punctuation
    for i in range (len(punctuation)):
        source= source.replace(punctuation[i]," ")
        
    return source

#불용어 제거 함수
def stopword_delete(words):
    stopword = [ 'A','a', 'and', 'are', 'as', 'at', 'be', 'but','But', 'by', 'for', 'if','If', 'in', 'into', 'is', 'it','It', 'no','No', 'not','Not', 'of', 'on', 'or', 's', 'such', 't', 'that','That', 'the','The', 'their','Their', 'then','Then', 'there','There', 'these','These', 'they','They', 'this','This', 'to', 'was', 'will', 'with']
    for i in range (len(stopword)):
        for j in range(len(words)):
            if words[j] == stopword[i]:
                words[j] = " "


#사전형태로 출현빈도 저장 함수
def word_count(words):
    words_dict = {}
    for i in words:
        if words_dict.has_key(i) is False:
            words_dict[i]=1
        else:
            words_dict[i]+=1
    return words_dict

def MakeDic(source):
    source=stylesheet_delete(source)
    source=javascript_delete(source)
    source=tag_delete(source)
    source=punc_delete(source)
    words=source.split()
    stopword_delete(words)
    source=" ".join(words)
    words=source.split()
    words=word_count(words)
    return words

class WebWordsFrequency():
    def __init__(self, *arg):
        self.URL = []
        for url in arg:
            self.URL.append(url)
    def addUrl(self, url):
        self.URL.append(url)
    def removeUrl(self, url):
        if url in self.URL:
            self.URL.remove(url)
    def listUrls(self):
        for url in self.URL:
            print url
    def getWordsFrequency(self):
        combine={}
        temp = []
        count = 0
        for url in self.URL:
            temp = MakeDic(urllib2.urlopen(url).read())
            count=1
            if(count==0):
                for i in temp:
                    combine[i]=temp[i]
            else:
                for i in temp:
                    if combine.has_key(i):
                        combine[i]+=temp[i]
                    else:
                        combine[i]=temp[i]
        return combine
    def getMaxFreqencyWords(self):
        combine = self.getWordsFrequency()
        g = combine.items()
        g.sort(key=lambda x:x[1],reverse=True)
        wordlist = []
        for i in range(len(g)-1):
            wordlist.append(g[i])
            if(g[i][1]!=g[i+1][1]):
                break
        if(len(wordlist)==0):
            return None
        return wordlist
#2확인
w1 = WebWordsFrequency('http://www.bestbuy.com/', 'http://www.times.com', 'https://www.amazon.com')
w2 = WebWordsFrequency('http://www.cnn.com', 'http://www.times.com')
w3 = WebWordsFrequency()
#3확인
w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')
#4확인
w1.removeUrl('http://www.cnn.com')
w2.removeUrl('http://stackoverflow.com')
#5확인
w1.listUrls()
#6확인
print w1.getWordsFrequency() 
#7확인
w=WebWordsFrequency()
print w1.getMaxFreqencyWords()
print w2.getMaxFreqencyWords()
```
